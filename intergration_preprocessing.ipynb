# TÍCH HỢP THÀNH FILE CSV

```python
import os
import json
import pandas as pd
import re

root_folder = "E:\INTERN\PORTFOLIO\AUTOMATION\CRAWL\data"
output_csv = "data/merge_data.csv"

def load_json_files(folder_path):
    data_list = []
    
    for website_folder in os.listdir(folder_path):
        website_path = os.path.join(folder_path, website_folder)
        if os.path.isdir(website_path): 
            for subdir, _, files in os.walk(website_path):
                for file_name in files:
                    if file_name == "data.json":
                        file_path = os.path.join(subdir, file_name)
                        with open(file_path, "r", encoding="utf-8") as f:
                            try:
                                data = json.load(f)
                                standardized_data = {
                                    "website": website_folder, 
                                    "scrape date": data.get("scrape_date", None),
                                    "sku": data.get("sku", None),
                                    "name": data.get("name", None),
                                    "sale": data.get("sale", None),
                                    "old_price": data.get("old_price", None),
                                    "special_price": data.get("special_price", None),
                                    "sizes": data.get("sizes", None),
                                    "colors": data.get("colors", None),
                                    "description": data.get("description", None),
                                    "images": data.get("images", None),
                                    "path": file_path  
                                }
                                data_list.append(standardized_data)
                            except json.JSONDecodeError:
                                print(f"Lỗi khi đọc file {file_name} trong thư mục {subdir}")
    
    return data_list

def merge_json_to_csv(root_folder, output_csv):
    data_list = load_json_files(root_folder)
    
    if not data_list:
        print("Không có dữ liệu để xuất.")
        return
    df = pd.DataFrame(data_list)
    
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")
    print(f"Dữ liệu đã được lưu vào {output_csv}")

merge_json_to_csv(root_folder, output_csv)
```

## XỬ LÝ THEO CỘT

```python
import pandas as pd
import re

# Đọc dữ liệu từ file CSV
df = pd.read_csv(r'E:\INTERN\PORTFOLIO\AUTOMATION\SYSTEM\data\merge_data.csv')
df = df.dropna(subset=['sku', 'name'], how='all')

# Xử lý cột 'sku':
df['sku'] = df['sku'].astype(str).str.replace("SKU", "").str.strip()

# Tìm các dòng có 'sku' bị null hoặc rỗng
mask_sku_null = df["sku"].isna() | (df["sku"] == "") | (df["sku"].str.lower() == "nan")

# Biểu thức regex để tìm SKU trong cột 'name'
pattern = r"\b[A-Z]+\d+\b"

def extract_sku(name):
    if pd.isna(name):  
        return None
    match = re.search(pattern, name)  # Tìm mã SKU trong name
    return match.group(0) if match else None

# Điền SKU cho các dòng bị thiếu
df.loc[mask_sku_null, "sku"] = df.loc[mask_sku_null, "name"].apply(extract_sku)

# Xóa SKU khỏi 'name' và dấu '-' ngay sau SKU
def clean_name(name):
    if pd.isna(name):
        return name
    return re.sub(r"\b[A-Z]+\d+\b\s*-*\s*", "", name).strip()

df["name"] = df["name"].apply(clean_name)

# Xử lý cột 'sale':
# - Loại bỏ ký tự '-' và '%', sau đó chuyển về float
# - Nếu giá trị là NaN, thay bằng 0
df["sale"] = df["sale"].astype(str).str.replace("-", "").str.replace("%", "").replace("", "0").astype(float) / 100
df["sale"].fillna(0, inplace=True)

# Xử lý cột 'old_price':
# - Loại bỏ ký tự không phải số, dấu phẩy và chữ cái, chỉ giữ lại số
# - Nếu giá trị là NaN, thay bằng 0
df["old_price"] = df["old_price"].astype(str).apply(lambda x: re.sub(r'[^0-9]', '', x) if pd.notna(x) else "0").replace("", "0").astype(float)
df["old_price"].fillna(0, inplace=True)

# Xử lý cột 'special_price' (nếu có):
# - Loại bỏ ký tự không phải số, chuyển đổi về số
# - Nếu giá trị rỗng hoặc NaN, thay bằng 0
if "special_price" in df.columns:
    df["special_price"] = df["special_price"].astype(str).apply(lambda x: re.sub(r'[^0-9]', '', x) if pd.notna(x) else "0").replace("", "0").astype(float)
    df["special_price"].fillna(0, inplace=True)

# Nếu cột 'sale' có giá trị nhưng 'special_price' bị null thì tính lại giá trị của 'special_price'
df.loc[(df["sale"] > 0) & (df["special_price"] == 0), "special_price"] = df["old_price"] - (df["old_price"] * df["sale"])

# Nếu 'special_price' có giá trị nhưng 'sale' là 0 thì tính lại giá trị 'sale'
df.loc[(df["special_price"] > 0) & (df["sale"] == 0), "sale"] = (df["old_price"] - df["special_price"]) / df["old_price"]

# Chuẩn hóa cột "size"
df["sizes"] = df["sizes"].astype(str)  # Đảm bảo kiểu dữ liệu là chuỗi
df["sizes"] = df["sizes"].str.replace(r"[\[\]']", "", regex=True)  # Xóa [] và ''
df["sizes"] = df["sizes"].str.replace(",", "_")  # Đổi dấu "," thành "_"
df["sizes"] = df["sizes"].str.replace(" ", "")  # Xóa dấu cách


# Chuẩn hóa cột "colors"
df["colors"] = df["colors"].astype(str)  # Đảm bảo kiểu dữ liệu là chuỗi
df["colors"] = df["colors"].str.replace(r"[\[\]']", "", regex=True)  # Xóa [] và ''
df["colors"] = df["colors"].str.replace(",", "_")  # Đổi dấu "," thành "_"
df["colors"] = df["colors"].str.replace(" ", "")  # Xóa dấu cách
df["colors"] = df["colors"].str.replace("Màusắc:", "") 


# Chuẩn hóa cột "description"
df["description"] = df["description"].astype(str)  # Đảm bảo kiểu dữ liệu là chuỗi
df["description"] = df["description"].str.replace(r"[\[\]']", "", regex=True)  # Xóa [] và ''
df["description"] = df["description"].str.replace(",", "_")  # Đổi dấu "," thành "_"


# Lưu dữ liệu đã xử lý vào file Excel
df.to_csv(r'E:\INTERN\PORTFOLIO\AUTOMATION\SYSTEM\data\clean_data.csv', index=False)

```
